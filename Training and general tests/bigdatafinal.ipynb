```
from pyspark.sql import SparkSession

spark = SparkSession\
  .builder\
  .appName("app_name")\
  .getOrCreate()
```

```
text_file = sc.textFile("dbfs:/FileStore/shared_uploads/fabrizio.flores@ucsp.edu.pe/pg10.txt")
counts = text_file.flatMap(lambda line: line.split(" ")) \
             .map(lambda word: (word, 1)) \
             .reduceByKey(lambda a, b: a + b)
counts.saveAsTextFile("dbfs:/FileStore/shared_uploads/fabrizio.flores@ucsp.edu.pe/123.txt")
```
